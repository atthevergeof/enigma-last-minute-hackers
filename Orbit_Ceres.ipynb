```python
import rebound
import matplotlib
from math import sin
from math import pi
```

```python
e=0.07610292126891821
M_deg = 340.38908482126703*pi/180
inc = 10.60069567603618*pi/180
f = M_deg + (2*e - 0.25 * e**3)*sin(M_deg) + (5/4)*(e**2)*sin(2*M_deg) + (13/12)*(e**3)*sin(3*M_deg)
Omega = 80.65851514365535*pi/180
omega = 71.4492152612411 * pi/180
```

```python
sim = rebound.Simulation()
sim.add(m=1)
sim.add(m=4.5037707*(10**(-10)), e=0.07610292126891821, a=2.767218108, inc=inc, f=f, Omega=Omega, omega=omega)
```

```python
%matplotlib inline
fig = rebound.OrbitPlot(sim)
```
